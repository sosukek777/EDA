```python
plt.plot(np.log(s))
```

```python
def rankappr(mat, r):
    u, s, v = np.linalg.svd(df, full_matrices=True, compute_uv=True)
    u_r = u[:, :r]
    s_r = diagsvd(s, 359, 371)[:r, :r]
    v_r = v[:r, :]
    
    return np.dot(u_r, np.dot(s_r, v_r))
```

### (i) $r = 1$

```python
plt.imshow(rankappr(df, 1), cmap='gray', vmin=1, vmax=64)
```

### (ii) $r = 20$

```python
plt.imshow(rankappr(df, 20), cmap='gray', vmin=1, vmax=64)
```

### (iii) $r = 50$

```python
plt.imshow(rankappr(df, 50), cmap='gray', vmin=1, vmax=64)
```

### (iv) $r = 100$

```python
plt.imshow(rankappr(df, 100), cmap='gray', vmin=1, vmax=64)
```
